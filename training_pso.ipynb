```python
import time
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers.legacy import Adam
from tensorflow.keras.losses import Hinge
from pyswarm import pso
import numpy as np
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
import numpy as np
from scikeras.wrappers import KerasClassifier
```

```python
start_time = time.time()
# Set your data directory
data_directory = '/Users/shreyas/Desktop/5th_sem/DeepLearning_Lab/Project/training_experimental/data/train'  # Update this path to the directory containing your dataset

# Define constants
input_shape = (48, 48, 1)  # Updated input size to match your image size
num_classes = 7  # Automatically determine the number of classes
epochs = 50

# Data preprocessing and augmentation
datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,  # Normalize pixel values to the [0, 1] range
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load and preprocess the data
datagen_with_preprocessing = datagen.flow_from_directory(
    data_directory,
    target_size=(48, 48),  # Updated target size
    color_mode='grayscale',
    batch_size=8192,
    class_mode='categorical'
)

# Split the data into training and validation sets
X_train, y_train = datagen_with_preprocessing.next()
X_val, y_val = datagen_with_preprocessing.next()

# Define the hyperparameter optimization objective function
def optimize_hyperparameters(params):
    learning_rate, batch_size = params
    batch_size = int(round(batch_size))

    # Create and compile the CNN model
    model = keras.Sequential([
        layers.Conv2D(64, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dropout(0.2),
        layers.Dense(512, activation='relu'),
        layers.Dense(num_classes, activation='softmax')  # Ensure num_classes matches your dataset
    ])

    optimizer = Adam(learning_rate=learning_rate)
    loss = 'categorical_crossentropy'  # You can change the loss function here

    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

    # Train the model
    history = model.fit(
        X_train,
        y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )

    # Calculate the test accuracy
    test_loss, test_accuracy = model.evaluate(X_val, y_val, verbose=0)
    
    # Minimize the negative of accuracy as PySwarm minimizes the objective function
    return -test_accuracy

# Define the parameter ranges for learning rate and batch size
lb = [0.00001, 32]
ub = [0.1, 8192]

# Use PSO to find optimal hyperparameters
best_hyperparameters, _ = pso(optimize_hyperparameters, lb, ub, swarmsize=20, maxiter=5)

# Retrieve the best learning rate and batch size
best_learning_rate, best_batch_size = best_hyperparameters

print("Best Learning Rate:", best_learning_rate)
print("Best Batch Size:", best_batch_size)
end_time = time.time()
elapsed_time = end_time - start_time
print("Elapsed Time: ", elapsed_time)
```
